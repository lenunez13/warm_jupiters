```python
%matplotlib inline
import rebound
import numpy as np
import matplotlib.pyplot as plt
import reboundx
from reboundx import constants
Mj = 955e-6 #...Jupiter mass in Solar units
```

```python
sim = rebound.Simulation()
sim.add(["Sun","Mercury","Venus","Earth","Mars","Jupiter","Saturn","Uranus","Neptune"],date="2018-02-10 00:00")
sim.save("ss.bin")
```

```python
sim = rebound.Simulation.from_file("ss.bin")
fig1 = rebound.OrbitPlot(sim,unitlabel="[AU]", color='blue', periastron=True, glow=True,fancy=False,slices=False);
```
