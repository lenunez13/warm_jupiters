````python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import rebound
import reboundx
import numpy as np
````

````python
#...Get eccentricity results from Mercury:

path = 'mercury' #/gpfs/group/rxd44/default/len56/mercury/
if os.path.exists(path+'/plan1.aei'):
    os.rename(path+'/plan1.aei',path+'/plan1.txt')
f = open(path+'/plan1.txt', 'r').readlines()[4:]
Nout = len(f)
eccentricity_mer = np.zeros(Nout)
times_mer = np.zeros(Nout)
for i in range(0,Nout):
    w = f[i].split()
    times_mer[i] = w[0]
    eccentricity_mer[i] = w[2]
````

````python
sim = rebound.Simulation()
sim.add(m=1)
sim.add(m=0,a=0.2,e=0.1,inc=np.radians(45.0))
ps = sim.particles

rebx = reboundx.Extras(sim)
mod = rebx.load_operator("modify_orbits_direct2")
rebx.add_operator(mod)

days = 2.0*np.pi/365.25
tmax = -500*days
ps[1].params["tau_e"] = tmax
````

````python
times = times_mer*days
Nout = len(times)
e1= np.zeros(Nout)
sim.integrator='ias15'
sim.ri_ias15.epsilon=0 #...to be used with fixed timestep in IAS
for i,time in enumerate(times):
    sim.integrate(time)
    e1[i] = ps[1].e
````

````python
fig = plt.figure(figsize=(15,5))
ax = plt.subplot(111)
plt.plot(times/days,e1,label='REBOUND',linewidth=3,color='black')
plt.plot(times/days,eccentricity_mer,label='Mercury',linewidth=3,linestyle='--',color='cyan')
plt.axes().set_xlabel("Time (Myr)", fontsize=18)
plt.axes().set_ylabel("Eccentricity", fontsize=18)
plt.legend(fontsize=18)

dir_name = 'plots/'
test = os.listdir(dir_name)

for item in test:
    if item.endswith('.pdf'):
        os.remove(os.path.join(dir_name, item))

savetag = 'orbital_evolution'
savetype = '.pdf'
#plt.savefig('plots/'+savetag+savetype,dpi=200,bbox_inches='tight')
````
