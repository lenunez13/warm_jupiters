```python
%matplotlib inline
import rebound
import numpy as np
import matplotlib.pyplot as plt
import reboundx
from reboundx import constants
```

```python
sim = rebound.Simulation()
sim.add(m=1., hash="star") # Sun
sim.add(m=1.66013e-07,a=0.387098,e=0.205630, hash="planet") # Mercury-like
sim.move_to_com() # Moves to the center of momentum frame
#rebound.OrbitPlot(sim, unitlabel="[AU]", color=False, periastron=True, glow=True,fancy=False,slices=False);

N = 1000
nyears = 1000
times = np.linspace(0.,2.*np.pi*nyears,N)
f = np.zeros(N)
pomega = np.zeros(N)
rebx = reboundx.Extras(sim)
gr = rebx.add("gr")
gr.params["c"] = constants.C
for i,t in enumerate(times):
    sim.integrate(t,exact_finish_time=0)
    f[i] = sim.particles[1].f
    pomega[i] = sim.particles[1].pomega
    
plt.figure(figsize=(10,5))
plt.plot(times/(2*np.pi),pomega)
```
