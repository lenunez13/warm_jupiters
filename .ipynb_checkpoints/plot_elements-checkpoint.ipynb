```python
%matplotlib inline
import rebound
import numpy as np
import matplotlib.pyplot as plt
import reboundx
from reboundx import constants
```

```python
Mj = 955e-6 #...Jupiter mass in Solar units


sim = rebound.Simulation()
sim.add(m=0.927)
sim.add(a=0.239,e=0.469,m=2.127*Mj,omega=66.*np.pi/180.0,inc=35.6*np.pi/180,Omega=0.)
sim.add(a=1.92,e=0.133,m=6.59*Mj,omega=136.9*np.pi/180.0,inc=3.4*np.pi/180,Omega=180.*np.pi/180.)
sim.move_to_com()
#fig = rebound.OrbitPlot(sim, unitlabel="[AU]", color=True, periastron=True)
rebx = reboundx.Extras(sim)
#gr = rebx.add("gr")
gr = rebx.add("gr_full")
gr.params["c"] = constants.C
```

```python
sim.dt = sim.particles[1].P/60. # small fraction of inner planet's period
sim.integrator="mercurius"
N = 1000
times = np.linspace(0.,2.*np.pi*0.05*1e6,N)

e1 = np.zeros(N)
inc1 = np.zeros(N)
theta1 = np.zeros(N)
f1 = np.zeros(N)


e2 = np.zeros(N)
inc2 = np.zeros(N)
theta2 = np.zeros(N)
f2 = np.zeros(N)

#...Simulation for Planet 1
for i,t in enumerate(times):
    sim.integrate(t,exact_finish_time=0)
    orbit = sim.particles[1].calculate_orbit(primary=sim.particles[0])
    e1[i] = orbit.e
    inc1[i] = orbit.inc
    theta1[i] = orbit.theta
    f1[i] = orbit.f

#...Simulation for Planet 2
for i,t in enumerate(times):
    sim.integrate(t,exact_finish_time=0)
    orbit = sim.particles[2].calculate_orbit(primary=sim.particles[0])
    e2[i] = orbit.e
    inc2[i] = orbit.inc
    theta2[i] = orbit.theta
    f2[i] = orbit.f

```

```python
fig = plt.figure(figsize=(9,7))
ax = plt.subplot(111)
ax.set_xlim([0,np.max(times)/2./np.pi])
ax.set_xlabel("time [yrs]")
ax.set_ylabel("eccentricity")
plt.plot(times/2./np.pi,e1);

fig = plt.figure(figsize=(9,7))
ax = plt.subplot(111)
ax.set_xlim([0,np.max(times)/2./np.pi])
ax.set_xlabel("time [yrs]")
ax.set_ylabel("Inclination")
plt.plot(times/2./np.pi,np.degrees(inc1-inc2));

fig = plt.figure(figsize=(9,7))
ax = plt.subplot(111)
ax.set_xlim([0,np.max(times)/2./np.pi])
ax.set_xlabel("time [yrs]")
ax.set_ylabel("Apsidal separation")
plt.plot(times/2./np.pi,np.degrees((theta1-f1) - (theta2-f2)));

```
