# Setting Up Simulations of Saturn's Moons

Based off of rebound/ipython_examples/Resonances_of_Jupiters_moons.ipynb
Now includes additional grav harmonics for central mass (due to Saturn's oblateness, etc)
Modifications based off of reboundx/ipython_examples/J2.ipynb

Let us first import REBOUND, numpy and matplotlib. We then download the current coordinates for Saturn and its moons from the NASA HORIZONS database. We work in units of km, days and solar masses.

```python
import rebound #for this to work correctly, need to install rebound on computer this is running on...
#to do that see instructions here: https://github.com/hannorein/rebound for installation with Python
#after rebound is installed, put this file in a directory under the reboud directory such as in the
#'ipython_examples directory or a directory you create
import reboundx #package for rebound to add additional forces (see https://github.com/dtamayo/reboundx for source code)
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt


sim = rebound.Simulation()
sim.units = ('km', 'days', 'kg')

# We can add Saturn and four of its moons by name, since REBOUND is linked to the HORIZONS database.
labels = ["Saturn", "Mimas", "Enceladus","Tethys","Dione","Rhea","Titan","Iapetus","Hyperion","Telesto","Calypso","Helene","Polydeuces"]
#labels = ["Saturn", "Mimas", "Enceladus","Tethys","Dione","Rhea","Titan","Iapetus","Hyperion"] #for troubleshooting without Trojan moons
#sim.add(labels)#,date="1980-11-12 23:47") #picking standard date to use

radius=[1.,2.,3.,4.,5.]
initdata=np.loadtxt('Saturnsats201706011200.txt',delimiter=', ') #this data is JPL Horizons
for i in xrange(len(labels)):
    if i==0:
        sim.add(m=5.68483446209e+26) #mass of saturn in kg
    else:
        sim.add(m=initdata[i-1,0],r=radius[i],x=initdata[i-1,1],y=initdata[i-1,2],z=initdata[i-1,3],vx=initdata[i-1,4],vy=initdata[i-1,5],vz=initdata[i-1,6])
#sim.add(labels,date="2017-06-01 12:00",plane="ecliptic")#,plane="500@6") #picking standard date to use
#coorbitals have no mass entry in horizons.py --> might need to modify later once we are simulating E ring particles...
#from JPL horizons:

#Polydeuces has no mass entry in JPL horizons
sim.status()
```

```python
#Now adding affects of Saturn's oblateness etc. (higher order gravitational harmonics)
#sim.move_to_com()
ps = sim.particles
#ps[4].m=7.194e15 #mass telesto in kg from JPL Horizons
# ps[4].m=4.7947258e15 #mass telesto in kg from  Dobrovolskis et al. (2010) assuming 0.6 g/cc density
#ps[5].m=3.597e15 #mass calypso in kg from JPL Horizons
# ps[5].m=2.9967e15 #mass calypso in kg from  Dobrovolskis et al. (2010) assuming 0.6 g/cc density
#ps[7].m=2.547e16 #mass helene in kg from JPL Horizons
#ps[7].m=1.0937968e16 #mass helene in kg from  Dobrovolskis et al. (2010) assuming 0.6 g/cc density
# ps[8].m=5.5439017e12 #mass polydeuces in kg from  Dobrovolskis et al. (2010) assuming 0.6 g/cc density
rebx = reboundx.Extras(sim)
rebx.add("gravitational_harmonics")
#values for Saturn params below are from Jacobson et al. 2006
# ps[0].params["J2"] = 1.629071e-2 #these are defining the 1st 'particle' (i.e. Saturn) to have these grav harmonic attributes
# ps[0].params["J4"] = -9.3583e-4
# ps[0].params["J6"] = 8.614e-5 #current rebx implementation does not include J6 or above
#values from  Dobrovolskis et al. (2010) (originally from Nicholson  Porco 1993:
# ps[0].params["J2"] = 1.6297e-2 #these are defining the 1st 'particle' (i.e. Saturn) to have these grav harmonic attributes
# ps[0].params["J4"] = -9.1e-4
#ps[0].params["J6"] = 1.070e-2 #current rebx implementation does not include J6 or above
#values from  Null et al. (1981):
# ps[0].params["J2"] = 1.6299e-2 #these are defining the 1st 'particle' (i.e. Saturn) to have these grav harmonic attributes
# ps[0].params["J4"] = -9.16e-4
# ps[0].params["J6"] = 8.1e-5

#values from  JPL Horizons (https://ssd.jpl.nasa.gov/?gravity_fields_op), orbit determinations use up to J8:
ps[0].params["J2"] = 1.629071e-2 #these are defining the 1st 'particle' (i.e. Saturn) to have these grav harmonic attributes
ps[0].params["J4"] = -9.3583e-4
#ps[0].params["J6"] = 8.614e-5
#radius value (radius at 1 bar atm pressure) noted below from JPL Horizons
#ps[0].params["R_eq"] = (6.0268e4) #equatorial radius in km  from JPL Horizons[these are the length units we defined previously]
ps[0].params["R_eq"] = (6.0330e4) #equatorial radius in km used for grav harmonics
#add below to energy sum
#rebx.gravitational_harmonics_hamiltonian(sim)
#see https://github.com/dtamayo/reboundx/blob/master/doc/effects.rst for documentation
sim.move_to_com()
```

Let us now calculate the mean motions and periods of the moons out to Polydeuces

```python
os = sim.calculate_orbits()#primary=ps[0])
# print("n_i (in rad/days) = %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f" % (os[0].n,os[1].n,os[2].n,os[3].n,os[4].n,os[5].n,os[6].n,os[7].n))
# print("P_i (in days)     = %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f" % (os[0].P,os[1].P,os[2].P,os[3].P,os[4].P,os[5].P,os[6].P,os[7].P))
# print("a_i (in km)     = %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f" % (os[0].a,os[1].a,os[2].a,os[3].a,os[4].a,os[5].a,os[6].a,os[7].a))

print("P_i (in days)     = %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f" % (os[0].P,os[1].P,os[2].P,os[3].P,os[4].P,os[5].P))
print("ecc     = %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f" % (os[0].e,os[1].e,os[2].e,os[3].e,os[4].e,os[5].e))
#print("inc     = %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f" % (os[0].i,os[1].i,os[2].i,os[3].i,os[4].i,os[5].i))
#print(ps[:].m)
print(ps[0].m,ps[1].m,ps[2].m,ps[3].m,ps[4].m,ps[5].m,ps[6].m)#,ps[7].m)#,ps[8].m,ps[9].m,ps[10].m)#,ps[11].m)
#print("mass (kg)     = %6.3f, %6.3f, %6.3f, %6.3f, %6.3f, %6.3f" % (os[0].m,os[1].m,os[2].m,os[3].m,os[4].m,os[5].m))
#print("Saturn mass:",os[0].)
print(sim.units)
```

We can see that some of the moons have similar mean motions (and orbital periods)...these are the moons orbiting at the L4 and L5 Lagrange points of Tethys and Dione
Enceladus and Dione should be in a 2:1 mean motion resonance with each other (so are Mimas and Tethys, but that case is slightly different due to the orbit inclinations so we will focus on Enceladus and Dione for now)

Let's move to the center of mass (COM) frame and plot the orbits of the these moons around Saturn:

```python
#sim.move_to_com()
fig = rebound.OrbitPlot(sim, unitlabel="[km]", color=True, trails=True, periastron=True)
```

We can now start integrating the system forward in time. This example uses the symplectic Wisdom-Holman type `whfast` integrator since no close encounters are expected. The timestep is set to 5% of one of Mimas' orbits.

```python
sim.integrator = "ias15" #change to ias15 when integrating test particles (Enceladus ejecta) in the system...
#sim.ri_mercurius.rcrit = 5
sim.dt = 0.05 * os[0].P  # 5% of Mimas's period
Nout = 100000            # number of points to display
tmax = 80*365.25# *24*3600        # let the simulation run for 80 (Earth) years
#Nmoons=6
Nmoons = 10 #number of moons in sim #6 when not including co-orbitals
Nmoons = 12 #number of moons in sim with Iapetus and Hyperion
Nmoons = 8 #number of moons in sim with Iapetus and Hyperion without co-orbs
print(sim.N)
Nmoons=sim.N-1
```

Similar to as was done in the Fourier analysis & resonances example, we set up several arrays to hold values as the simulation runs.  This includes the positions of the moons, eccentricities, mean longitudes, and longitude of pericentres.

```python
x = np.zeros((Nmoons,Nout))
y = np.zeros((Nmoons,Nout))
z = np.zeros((Nmoons,Nout))
ecc = np.zeros((Nmoons,Nout))
pd= np.zeros((Nmoons,Nout))
longitude = np.zeros((Nmoons,Nout))
varpi = np.zeros((Nmoons,Nout))
Om = np.zeros((Nmoons,Nout))
longperi = np.zeros((Nmoons,Nout))
sime= np.zeros(Nout)
times = np.linspace(0.,tmax,Nout)
#ps = sim.particles #when adding reboundx, I define this earlier further up in the code
#THIS IS A BIG NUMBER CRUNCH STEP...MIGHT TAKE A FEW MINUTES
for i,time in enumerate(times):
    sim.integrate(time,exact_finish_time=0)
    # default is exact_finish_time=1, which changes the integration timestep near 
    # the outputs to match the output times we want. This 
    # technically breaks WHFast's symplectic nature...let's try to avoid that for better accuracy
    #sim.move_to_com()
    os = sim.calculate_orbits()#primary=ps[0])
    sime[i]=sim.calculate_energy()+rebx.gravitational_harmonics_hamiltonian(sim) #calculates total energy for all particles in the simulation at each timestep
    for j in range(Nmoons): #recording data of interest for each moon
        x[j][i] = ps[j+1].x #x coordinates of moons
        y[j][i] = ps[j+1].y #y coordinates of moons
        z[j][i] = ps[j+1].z #z coordinates of moons
        ecc[j][i] = os[j].e #orbit eccentricity of each moon
        pd[j][i] = os[j].P #orbit eccentricity of each moon
        longitude[j][i] = os[j].l
        varpi[j][i] = os[j].Omega + os[j].omega
        Om[j][i] = os[j].Omega 
        longperi[j][i] =os[j].pomega
        
#Calculate max fractional energy change over simulation:
dE=(sime-sime[0])/sime[0]
print("Max dE/E:",np.max(np.abs(dE))) 
#dE should be a tiny number!!! Otherwise the simulation isn't very accurate
```

```python
fig = plt.figure(figsize=(12,5))
ax = plt.subplot(111)
#remember that index of data arrays is shifted from label array that includes Saturn
plt.plot(times,ecc[0],label=labels[1]) 
plt.plot(times,ecc[1],label=labels[2])
plt.plot(times,ecc[2],label=labels[3])
plt.plot(times,ecc[3],label=labels[4])
plt.plot(times,ecc[4],label=labels[5])
plt.plot(times,ecc[5],label=labels[6])
ax.set_xlabel("Time (days)")
ax.set_ylabel("Eccentricity")
plt.legend();
print("Min dE/E>0:",np.min(np.abs(dE[np.abs(dE)>0.]))) 
```

Things to look at later:
We can plot the Trojan (L4 and L5) moons' locations relative to Tethys and Dione as well.
And check whether the orbits of moons in mean motion resonances are evolving as we expect...

Resonances are identified by looking at the resonant arguments, the difference in relative orbital angles between two potentially resonant objects.  One type of mean motion resonance argument is defined as: 
   $$ \theta = (p + q)\lambda_{\rm out} - p \lambda_{\rm in} - q \varpi_{\rm out/in}$$
 where $\lambda_{\rm out}$ and $\lambda_{\rm in}$ are the mean longitudes of the outer and inner bodies, respectively,
 and $\varpi_{\rm out}$ is the longitude of pericenter of the outer/inner body.
 The ratio of periods is defined as : $$P_{\rm in}/P_{\rm out} ~= p / (p + q)$$

 If the resonant argument, $\theta$, oscillates but is constrained within some range of angles, then 
 there is a resonance between the inner and outer bodies. We call this libration of the angle $\theta$. 
 The trick is to find what the values of q and p are.  For our case, we can easily see that 
 there are two 2:1 resonances between the moons, so their resonant arguments would follow 
 the function:
    $$\theta = 2 \lambda_{\rm out} - \lambda_{\rm in} - \varpi_{\rm out}$$
 To make the plotting easier, we can borrow this helper function that puts angles into 0 to 360 degrees 
 from another example (Fourier analysis & resonances), and define a new one that puts angles
 into -180 to 180 degrees.

```python
def zeroTo360(val):
    while val < 0:
        val += 2*np.pi
    while val > 2*np.pi:
        val -= 2*np.pi
    return (val*180/np.pi)

def min180To180(val):
    while val < -np.pi:
        val += 2*np.pi
    while val > np.pi:
        val -= 2*np.pi
    return (val*180/np.pi)

def rad2deg(val): #this function converts angle in radians to degrees
    return (val*180/np.pi)

# We can calculate theta, the resonant argument of the 2:1 Enceladus-Dione orbital resonance,
# which oscillates (librates) about 0 degrees:
theta = [rad2deg(longitude[1][i] - 2*longitude[3][i] + varpi[1][i]) for i in range(Nout)]
#Mimas-Tethys res should librate with lib period of ~11yrs (Jacobson et al. 2006)
#res angle in sim without co-orbs:
#theta = [rad2deg(longitude[1][i] - 2*longitude[3][i] + varpi[1][i]) for i in range(Nout)]
# We can calculate theta, the resonant argument of the 2:1 Mimas-Tethys orbital resonance,
# which oscillates (librates) about 0 degrees:
theta2 = [rad2deg(2*longitude[0][i] - 4*longitude[2][i] + Om[2][i]+Om[0][i]) for i in range(Nout)]
#Mimas-Tethys res should librate with lib period of 70.78 yrs  and amplitude of 97.04 deg (Allan 1969)
#conj2=[rad2deg(2*longitude[0][i] - 4*longitude[2][i]) for i in range(Nout)]
# We can calculate theta, the resonant argument of the 4:3 Titan:Hyperion orbital resonance,
# which oscillates (librates) about 180 degrees:
theta3 = [rad2deg(3*longitude[-7][i] - 4*longitude[-5][i] + varpi[-5][i]) for i in range(Nout)]
encap = [zeroTo360(varpi[1][i]) for i in range(Nout)] #Enceladus' longitude over time
tethap = [zeroTo360(varpi[2][i]) for i in range(Nout)] #tethys' longitude over time
diap = [zeroTo360(varpi[3][i]) for i in range(Nout)] #Dione's longitude over time
hypap = [zeroTo360(varpi[-5][i]) for i in range(Nout)] #Hyperion's longitude of periapse over time
theta=np.array(theta)%360 #wrapping angles from 0 to 360 degrees
theta2=np.array(theta2)%360
theta3=np.array(theta3)%360
#print("Enceladus-Dione Period Ratio:",np.mean(pd[5][:]/pd[1][:]))
print("Enceladus-Dione Period Ratio:",np.mean(pd[3][:]/pd[1][:])) #without co-orbs
print("Mimas-Tethys Period Ratio:",np.mean(pd[2][:]/pd[0][:]))
print("Titan-Hyperion Period Ratio:",np.mean(pd[-5][:]/pd[-7][:]))
print("Mean Tethys Orbit Distance",np.mean(np.sqrt((x[2][:]**2.+y[2][:]**2.+z[2][:]**2.))))


fig = plt.figure(figsize=(12,5))
ax = plt.subplot(111)
ax.plot(times/365.25,theta,'b.')  #this should be librating!!! 
ax.set_xlim([0,80.])
#ax.set_ylim([-180,180.])
ax.set_ylim([0,360])
ax.set_xlabel("Time (years)", fontsize=16)
ax.set_ylabel(r"Resonant Argument $\theta_{2:1}$", fontsize=16)
ax.set_title("Enceladus:Dione 2:1 MMR",fontsize=16)
ax.tick_params(labelsize=14)
#ax.plot([0,100],[180,180],'k--')

fig = plt.figure(figsize=(12,5))
ax = plt.subplot(111)
ax.plot(times/365.25,theta2,'b.')  
ax.set_xlim([0,80.])
#ax.set_ylim([-180,180.])
ax.set_ylim([0,360])
ax.set_xlabel("Time (years)", fontsize=16)
ax.set_ylabel(r"Resonant Argument $\theta_{2:1}$", fontsize=16)
ax.set_title("Mimas:Tethys 2:1 MMR",fontsize=16)
ax.tick_params(labelsize=14)
#Mimas-Tethys res arg circulates when including only up to J4 and estimated coorbital masses
#librates if using outdated J2 and J4

# fig = plt.figure(figsize=(12,5))
# ax = plt.subplot(111)
# ax.plot(times/365.25,conj2,'b.')  
# ax.set_xlim([0,80.])
# #ax.set_ylim([-180,180.])
# ax.set_ylim([0,360])
# ax.set_xlabel("Time (years)", fontsize=16)
# ax.set_ylabel(r"Conjunction", fontsize=16)
# ax.set_title("Mimas:Tethys 2:1 MMR",fontsize=16)
# ax.tick_params(labelsize=14)


fig = plt.figure(figsize=(12,5))
ax = plt.subplot(111)
ax.plot(times/365.25,theta3,'b.')  

ax.set_xlim([0,80.])
ax.set_ylim([0,360])
ax.set_xlabel("Time (years)", fontsize=16)
ax.set_ylabel(r"Resonant Argument $\theta_{4:3}$", fontsize=16)
ax.set_title("Titan:Hyperion 4:3 MMR",fontsize=16)
ax.tick_params(labelsize=14)

```
